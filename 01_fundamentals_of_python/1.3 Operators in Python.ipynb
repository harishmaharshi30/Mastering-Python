## Operators in Pyhton

**Definition:** 
- Operators in Python are special symbols or keywords used to perform operations on variables and values.
- They allow you to manipulate data by performing arithmetic, comparison, logical, or bitwise computations.
- Examples include +, -, *, /, ==, and, or, etc.

3. **Comparison Operator**

4. **Logical Operators**

5. Practical Exercise

```python
##  Arithmatic Operations
a = 10
b = 5
add_result = a+b           #addition Operations
sub_result = a-b           # Subraction Operations
mult_result = a*b          # Multiplication Operation
div_result = a/b           # divsion Operation
floor_div_result = a//b    # Floor Division
modulus_result = a % b     # Modules operation
exponential_result = a**b  # Exponential 
print(add_result)
print(sub_result)
print(mult_result)
print(div_result)
print(floor_div_result)
print(modulus_result)
print(exponential_result)
```
